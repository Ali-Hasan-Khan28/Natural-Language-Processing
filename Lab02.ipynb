```python
import pandas as pd
import nltk
from nltk import pos_tag, word_tokenize

# Download the necessary resource
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
# Recipes data
recipes_data = """
Chocolate Chip Cookies: Combine 2 cups of flour, 1 cup of sugar, 1 cup of brown sugar, 1 tsp of baking soda, and a pinch of salt. Cream together with 1 cup of softened butter, add 2 eggs and 2 tsp of vanilla extract. Fold in 2 cups of chocolate chips. Drop spoonfuls on a baking sheet and bake at 350°F for 10 minutes until golden brown.
"""

# Split the data into recipes
recipes_list = [recipe.strip() for recipe in recipes_data.split('\n') if recipe.strip()]

# Create a dataframe
df = pd.DataFrame(columns=['Recipe Name', 'Ingredients', 'Instructions', 'Measurements'])

# Function to determine whether a word is an ingredient, instruction, or measurement
def get_word_category(word):
    # Simple categorization based on part-of-speech tagging
    pos_tags = pos_tag([word])
    if pos_tags[0][1].startswith('NN'):  # NN stands for nouns
        return 'Ingredient'
    elif pos_tags[0][1] in ['VB', 'VBG', 'VBD', 'VBN']:  # VB stands for verbs
        return 'Instruction'
    elif pos_tags[0][1] in ['CD']:  # CD stands for cardinal numbers (measurements)
        return 'Measurement'
    else:
        return 'Other'

# Process each recipe
for recipe in recipes_list:
    recipe_info = recipe.split(':')
    recipe_name = recipe_info[0].strip()
    recipe_method = recipe_info[1].strip()

    # Tokenize words and categorize them
    words = word_tokenize(recipe_method)
    categorized_words = [(word, get_word_category(word)) for word in words]

    # Separate words into ingredients, instructions, and measurements
    ingredients = [word for word, category in categorized_words if category == 'Ingredient']
    instructions = [word for word, category in categorized_words if category == 'Instruction']
    measurements = [word for word, category in categorized_words if category == 'Measurement']

    # Append data to the dataframe
    df = df.append({'Recipe Name': recipe_name,
                    'Ingredients': ' '.join(ingredients),
                    'Instructions': ' '.join(instructions),
                    'Measurements': ' '.join(measurements)}, ignore_index=True)

# Display the resulting dataframe
print(df)
```
